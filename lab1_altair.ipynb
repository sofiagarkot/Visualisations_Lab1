```python
import altair as alt
import pandas as pd
import altair_viewer

alt.renderers.enable('altair_viewer', inline=True)
```

```python
df = pd.read_csv('GLB.Ts_dSST.csv', header=1)
df.head()
```

```python
type(df["Year"][0])
```

```python
df['q_Year'] = df["Year"] 
df["Year"] = df["Year"].astype("str")
df
```

```python
alt.Chart(df).mark_point(filled = True, color = '#db7209').encode(
    x = alt.X(field='Year', type = 'temporal',
             ),
    y = alt.Y(field = 'J-D', type = 'quantitative'), 
).properties(width = 800, height = 500, background = '#F9F9F9', padding = 25)


# alt.mark_circle(filled = False, color = 'black')
```

```python
circles = alt.Chart(df).mark_circle(filled = False, size = 55, color = '#373535', opacity=0.7).encode(
    x = alt.X(field='Year', type = 'temporal'),
    y = alt.Y(field = 'J-D', type = 'quantitative'), 
)

points = alt.Chart(df).mark_point(
    filled = True, size = 50, color='#f07043', opacity=1.0).encode(
    
    
    x = alt.X(field='Year', 
              type='temporal',
              axis = alt.Axis(tickCount = 10,
              labelColor='grey',title='',
              domain=False)
             ),
    
    y = alt.Y(field = 'J-D', 
              type = 'quantitative',
              axis = alt.Axis(tickCount = 9, 
              labelColor='grey',title='',
              domain=False,
              labelExpr='datum.value <-0.5 ? null : datum.label+"°"'), 
              scale=alt.Scale(domain=[-0.5, 1.2]))
    
)

line = alt.Chart(pd.DataFrame({'y': [0.0]})).mark_rule(color='grey').encode(y='y')

text_left = alt.Chart(df.query("Year in ['1904', '1998']")
    ).mark_text(dx=-20, color="black",font = 'Futura', fontWeight=400).encode(
    x=alt.X(field='Year', type='temporal'), 
    y=alt.Y(field = 'J-D',type = 'quantitative'), 
    text=alt.Text(field='q_Year', type='quantitative'))

text_right = alt.Chart(df.query("Year in ['1944', '2014', '2015', '2017']")
    ).mark_text(dx=20, color="black", font = 'Futura', fontWeight=400).encode(
    x=alt.X(field='Year', type='temporal'), 
    y=alt.Y(field = 'J-D',type = 'quantitative'), 
    text=alt.Text(field='q_Year', type='quantitative'))

text_bold = alt.Chart(df.query("Year in ['2016']")
    ).mark_text(dx=20, color="black", font = 'Futura',fontWeight='bold').encode(
    x=alt.X(field='Year', type='temporal'), 
    y=alt.Y(field = 'J-D',type = 'quantitative'), 
    text=alt.Text(field='q_Year', type='quantitative'))

text0 = alt.Chart({'values':[{'x': '2018', 'y': 0.1}]}).mark_text(
    text='HOTTER THAN THE\n1880 - 1899 AVERAGE', lineBreak='\n', font = 'Futura',
    align='right'
).encode(
    x='x:T', y='y:Q'
)

text1 = alt.Chart({'values':[{'x': '2018', 'y': -0.04}]}).mark_text(
    text='COLDER', font = 'Futura',
    align='right'
).encode(
    x='x:T', y='y:Q'
)


text_superscript_0 = alt.Chart({'values':[{'x': '1881', 'y': 1.2}]}).mark_text(
    text='Annual Global Surface Temperature,\nRelative to Late 19th Century Average', 
    align='left', font = 'Futura',
    fontSize=13, fontWeight='bold', color='#4B4545',  lineBreak='\n'
).encode(
    x='x:T', y='y:Q'
)



(circles+points+line+text_left+text_right+text_bold+text0+text1+text_superscript_0).properties(
    width = 800, height = 500, background = '#F9F9F9', padding = 25).configure_view(
    strokeWidth=0)
```
